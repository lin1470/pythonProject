```python
# 导入相关模块，读取数据
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filename = u'数据清理数据集.xls'
intro = pd.read_excel(filename, sheetname=0)
data = pd.read_excel(filename, sheetname=1, header=0,skiprows=[1,])

# 找出带’？'标志的特征
feature_with_question = []
for key in data.keys():
    try:
        if any(data[key]=='?'):
            feature_with_question += [key]
    except Exception:
        pass
print feature_with_question
```

### 上面的结果说明了在四个变量中有缺省值。

#### 1. 将数据补齐

共有四个特征存在缺省值：  
- BASE_POINTS_SUM  
- EXPENSE_SUM_YR_1  
- EXPENSE_SUM_YR_2  
- SEG_KM_SUM  


1. 由特征属性名推断，并观察数据，判定BASE_POINTS_SUM = P1Y_BASE_POINTS_SUM + L1Y_BASE_POINTS_SUM
2. unknown 暂时无法判断第一年总票价和其他特征之间的关系，认为无法推导出来，因此默认是0
3. unknown 暂时无法判断第二年总票价和其他特征之间的关系，认为无法推导出来，因此默认是0
4. 由特征属性名推断，并观察数据，判定WEIGHTED_SEG_KM = SEG_KM_SUM * avg_discount


```python
# 补全"?"标志的缺失数据
data['BASE_POINTS_SUM'] = data['P1Y_BASE_POINTS_SUM'] + data['L1Y_BASE_POINTS_SUM']
# 尽管不是缺省值，也可以重新覆盖掉的。
data['SEG_KM_SUM'] = data['WEIGHTED_SEG_KM'] / data['avg_discount']
for i in range(len(data['EXPENSE_SUM_YR_1'])): # 默认是缺省值是0
    if data['EXPENSE_SUM_YR_1'][i] == '?':
        data['EXPENSE_SUM_YR_1'][i] = 0
for i in range(1,len(data['EXPENSE_SUM_YR_2'])): # 默认是缺省值是0
    if data['EXPENSE_SUM_YR_2'][i] == '?':
        data['EXPENSE_SUM_YR_2'][i] = 0
```

#### 小题4: 将“观测窗口总飞行公里数”和“观测窗口总加权飞行公里数（Σ舱位折扣×航段距离）”属性对数据进行Z-Score规范化，并对两属性进行比较。

```python
def z_score_of(feature):
    mean = feature.mean()
    std = feature.std()
    return (feature-mean)/std

series_1 = z_score_of(data['SEG_KM_SUM'])
series_2 = z_score_of(data['WEIGHTED_SEG_KM'])

z_score_d = {'Z-SEG_KM_SUM': series_1, 'Z-WEIGHTED_SEG_KM': series_2}
z_score_df = pd.DataFrame(data=z_score_d)
data = data.join(z_score_df)

data.plot.scatter('Z-SEG_KM_SUM','Z-WEIGHTED_SEG_KM')
plt.show()
```
